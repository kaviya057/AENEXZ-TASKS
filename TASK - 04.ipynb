```python
import pandas as pd
df = pd.read_csv("instagram_data.csv")   
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['followers'])
```

```python
df['avg_likes_filled'] = df['avg_likes'].fillna(0)
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
import re

def clean_text(text):
    text = str(text).lower()                  
    text = re.sub(r'\d+', '', text)           
    text = re.sub(r'[^\w\s]', '', text)       
    text = re.sub(r'\s+', ' ', text).strip()  
    return text

df['clean_followers'] = df['followers'].apply(clean_text)
```

```python
df['followers'] = df['followers'].replace({
    "avg_likes": "likes"
})
```

```python
df.to_csv("cleaned_instagram_data.csv", index=False)
```
