```python
#Write a function that takes the radius of a circle and returns its area.

import math
def circle(r):
    return math.pi*r*r 
radius=int(input("enter radius to find area of a circle: "))
result=circle(radius)
print(result)
```

```python
#Write a function that accepts three numbers and returns the largest one.

def largest(a,b,c):
    if a<b and b>c:
        return b
    elif a>b and a>c:
        return a
    else:
        return c
a=int(input("enter number for a: "))
b=int(input("enter number for b: "))
c=int(input("enter number for c: "))
result=largest(a,b,c)
print(result)
```

```python
#Write a function that takes a string and returns its length without using built-in length functions.

def len_string(s):
    count=0
    for i in s:
        count+=1
    return count
s=input("enter a string: ")
result=len_string(s)
print(result)
```

```python
#Write a function that checks whether a number is prime.

def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
n=int(input("enter number for n: "))
result=is_prime(n)
print(result)
```

```python
#Write a recursive function to find the nth Fibonacci number.

def fibonacci(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fibonacci(n-1)+fibonacci(n-2)
n=int(input("enter number for n: "))
result=fibonacci(n)
print(result)
```

```python
#Write a recursive function to reverse a string.

def rev(s):
    if len(s)<=1:
        return s
    else:
        return rev(s[1:]) + s[0]
s=input("enter a string: ")
result=rev(s)
print(result)
```

```python
#Write a program to find the sum of all numbers from 1 to n using a for loop

def sum_of(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return sum_of(n-1)+n
n=int(input("enter number for n: "))
result=sum_of(n)
print(result)
```

```python
#Write a program to count how many even numbers are present in a list.

def even(lis):
    count=0
    for i in lis:
        if i%2==0:
            count+=1
    return count
a=list(map(int,input("enter a number: ").split()))
result=even(a)
print(result)
```
