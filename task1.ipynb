```python
#Write a program to find the largest of three numbers.

a=int(input("enter a number for 'a' value: "))
b=int(input("enter a number for 'b' value: "))
c=int(input("enter a number for 'c' value: "))
if a>b and a>c:
    print(a, "is largest of three numbers")
elif b>a and b>c:
    print(b, "is largest of three numbers")
else:
    print(c, "is largest of three numbers")
```

```python
#Write a program to check whether a given year is a leap year.

year=int(input("enter a year: "))
if (year%4==0 and year%100!=0) or year%400==0:
    print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

```python
#Write a program to check if a character is a vowel or consonant.

a=input("enter a letter: ")
if a in "aeiou" or "AEIOU":
    print(a,"is vowel")
else:
    print(a,"is consonant")
```

```python
#Write a program to calculate the sum of first N natural numbers using a while loop.

upto=int(input("enter any number: "))
i=1
sum=0
while i<=upto:
    sum+=i
    i+=1
print(sum,"sum of first",upto,"natural numbers")
```

```python
#Write a program to print the multiplication table of a given number using a while loop.

table=int(input("enter a number: "))
upto =int(input("enter a number: "))
i=1
while i<=upto:
    print(table,"*",i,"=",i*table)
    i+=1
```

```python
"""
Print a pyramid pattern:
    *
   ***
  *****
 *******
"""

size=int(input("enter a value: "))
for i in range(1,size+1):
    print(" "*(size-i),"*"*(2*i-1))
```

```python
"""
Print:
1
12
123
1234
12345
"""
size=int(input("enter a value: "))
for i in range(1,size+1):
    for j in range(1,i+1):
        print(j,end="")
    print()
```
